```python
import requests
from bs4 import BeautifulSoup

# URL of the IMDb Top Rated Movies page
url = "https://www.snapdeal.com/products/mens-footwear-sports-shoes?sort=plrty"

# Send a GET request to the URL
response = requests.get(url)
```

```python
response
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup
```

```python
titles = soup.find_all("p", class_="product-title")

```

```python
titles
```

```python
# Iterate over the movies and extract relevant information
title=[]
for i in titles:
  title.append(i.text)
```

```python
title
```

```python
rupee = soup.find_all("span", class_="lfloat product-price")
```

```python
rupee
```

```python
price=[]

for i in rupee:
  price.append(i.text)
```

```python
price
```

```python
discounts = soup.find_all("div", class_="product-discount")
```

```python
discounts
```

```python
discount=[]

for i in discounts:
  discount.append(i.text)
```

```python
discount
```

```python
reviews = soup.find_all("p", class_="product-rating-count")
```

```python
reviews
```

```python
no_reviews=[]

for i in reviews:
  no_reviews.append(i.text)
```

```python
no_reviews
```

```python
len(no_reviews),len(discount),len(price),len(title)
```

```python
from textblob import TextBlob

# Define a sample text
text = "UrbanMark Men Comfortable Contrast Panels Lace Up Training Sports Shoes - Navy"

# Create a TextBlob object
blob = TextBlob(text)

# Perform sentiment analysis
sentiment = blob.sentiment

# Print the sentiment polarity (-1 to 1, where negative is -1, positive is 1, and neutral is 0)
print("Polarity:", sentiment.polarity)

# Print the sentiment subjectivity (0 to 1, where 0 is very objective and 1 is very subjective)
print("Subjectivity:", sentiment.subjectivity)
```
